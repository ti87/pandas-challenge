```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df
```

## Player Count

* Display the total number of players


```python
player_demographics = purchase_data_df.loc[:, ["Gender", "SN", "Age"]]
player_demographics = player_demographics.drop_duplicates()
num_players = player_demographics.count()[0]
pd.DataFrame({"Total Players": [num_players]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
avg_price=purchase_data_df["Price"].mean()
total_purchases=len(purchase_data_df)
unique_items=len(purchase_data_df["Item ID"].unique())
total_revenue=purchase_data_df["Price"].sum()

summary_purchase_df=pd.DataFrame({"Number of Unique Items": [unique_items], "Average Price": [avg_price], 
                                 "Number of Purchases": [total_purchases], "Total Revenue": [total_revenue]})

summary_purchase_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_totals=player_demographics["Gender"].value_counts()
gender_percents=((gender_totals/num_players)*100)
gender_demographics_df=pd.DataFrame({"Total Count": gender_totals, "Percentage of Players": gender_percents})
gender_demographics_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_purchase_total = purchase_data_df.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")
gender_average = purchase_data_df.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Price")
gender_counts = purchase_data_df.groupby(["Gender"]).count()["Price"].rename("Purchase Count")

normalized_total = gender_purchase_total / gender_demographics_df["Total Count"]

gender_data_df = pd.DataFrame({"Purchase Count": gender_counts, "Average Purchase Price": gender_average, "Total Purchase Value": gender_purchase_total, "Normalized Totals": normalized_total})


gender_data_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
age_groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]


player_demographics["Age Ranges"] = pd.cut(player_demographics["Age"], age_bins, labels=age_groups)


age_demographics_totals = player_demographics["Age Ranges"].value_counts()
age_demographics_percents = age_demographics_totals / num_players
age_demographics = pd.DataFrame({"Total Count": age_demographics_totals, "Percentage of Players": age_demographics_percents})


age_demographics['Percentage of Players'] = age_demographics['Percentage of Players'].map("{:,.2%}".format)


age_demographics = age_demographics.sort_index()
age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data_df["Age Ranges"]=pd.cut(purchase_data_df["Age"], age_bins, labels=age_groups)

purchase_count=purchase_data_df.groupby(["Age Ranges"]).count()["Price"]
avg_purchase_price=purchase_data_df.groupby(["Age Ranges"]).mean()["Price"]
total_purchase=purchase_data_df.groupby(["Age Ranges"]).sum()["Price"]
avg_person_purchase=total_purchase/age_demographics["Total Count"]

age_data_df=pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": avg_purchase_price, 
                         "Total Purchase Value": total_purchase, "Avg Total Purchase per Person": avg_person_purchase})

age_data_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_purchase_count=purchase_data_df.groupby(["SN"]).count()["Price"]
top_avg_purchase_price=purchase_data_df.groupby(["SN"]).mean()["Price"]
top_total_purchase_value=purchase_data_df.groupby(["SN"]).sum()["Price"]

top_spenders_df=pd.DataFrame({"Purchase Count": top_purchase_count, "Average Purchase Price": top_avg_purchase_price, 
                             "Total Purchase Value": top_total_purchase_value})
top_spenders_df.sort_values("Total Purchase Value", ascending=False)
top_spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_info=purchase_data_df.loc[:,["Item ID", "Item Name", "Price"]]
item_purchase_count=item_info.groupby(["Item ID", "Item Name"]).count()["Price"]
avg_item_price=item_info.groupby(["Item ID", "Item Name"]).mean()["Price"]
total_item_purchase=item_info.groupby(["Item ID", "Item Name"]).sum()["Price"]

summary_item_df=pd.DataFrame({"Purchase Count": item_purchase_count, "Item Price": avg_item_price, 
                             "Total Purchase Value": total_item_purchase})
sorted_item_values=summary_item_df.sort_values("Purchase Count", ascending=False)
item_popularity=sorted_item_values.loc[:,["Purchase Count", "Item Price", "Total Purchase Value"]]
item_popularity.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting

```python
sorted_item_table=item_popularity.sort_values("Total Purchase Value", ascending=False)
sorted_item_table.head()
```
